1. Побудувати графіки для тригонометричних функцій, використати різні стилі та маркери

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, y1, linestyle='-', marker='o', markevery=20, label='sin(x)')
ax.plot(x, y2, linestyle='--', marker='s', markevery=20, label='cos(x)')

ax.set_title("y = sin (x), y = cos (x)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

2. Побудуйте точкову діаграму, де колір точки залежить від третьої змінної, додайте шкалу кольорів.

```python
np.random.seed(1)
x = np.random.randn(200)
y = 1.2 * x + np.random.randn(200) / 2

c = np.hypot(x, y)  # “інтенсивність” для кольору

fig, ax = plt.subplots(figsize=(6, 3))
sc = ax.scatter(x, y, c=c, s=50, edgecolor='k')
ax.set_title("Scatter: колір за величиною")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.colorbar(sc, ax=ax, label="шкала кольору")
ax.grid(True, alpha=0.3)

plt.show()
```

3. Побудуйте bar chart для 6 категорій і підпишіть значення над стовпчиками.

```python
cats = ["A", "B", "C", "D", "E", "F"]
vals = [12, 7, 13, 9, 11, 5]

fig, ax = plt.subplots(figsize=(6, 3))
bars = ax.bar(cats, vals)

ax.set_title("Bar chart із підписами")
ax.set_xlabel("Категорія")
ax.set_ylabel("Значення")
ax.grid(axis='y', alpha=0.3)

for b in bars:
    ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.3,
            f"{b.get_height():.0f}", ha="center", va="bottom")

plt.show()
```

4.Побудуйте гістограми двох розподілів на одному полі з легендою.

```python
np.random.seed(2)
a = np.random.normal(loc=0, scale=1, size=2000)
b = np.random.normal(loc=1, scale=1.2, size=2000)

fig, ax = plt.subplots(figsize=(6, 3))
ax.hist(a, bins=30, alpha=0.6, label="N(0,1)")
ax.hist(b, bins=30, alpha=0.6, label="N(1,1.2)")

ax.set_title("Дві гістограми")
ax.set_xlabel("Значення")
ax.set_ylabel("Частота")
ax.legend()
plt.show()
```

5. Побудуйте boxplot для 4 груп, підпишіть осі.

```python
np.random.seed(3)
data = [
    np.random.normal(0, 1, 200),
    np.random.normal(0.5, 1.1, 200),
    np.random.normal(1.0, 0.9, 200),
    np.random.normal(-0.2, 1.3, 200),
]

fig, ax = plt.subplots(figsize=(6, 3))
ax.boxplot(data, labels=["G1", "G2", "G3", "G4"], showmeans=True)

ax.set_title("Boxplot груп")
ax.set_xlabel("Група")
ax.set_ylabel("Значення")
ax.grid(axis="y", alpha=0.3)
plt.show()
```

6. Побудуйте дві криві та зафарбуйте область між ними.

```python
x = np.linspace(0, 8, 200)
y1 = 3 + 0.5 * x + 0.2 * np.sin(2*x)
y2 = 1 + 0.3 * x + 0.2 * np.cos(2*x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.fill_between(x, y1, y2, alpha=0.5)
ax.plot(x, (y1 + y2)/2, linewidth=2, label="Середня лінія")

ax.set_title("fill_between")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

7.Побудуйте кругову діаграму для 5 секторів із відсотками.

8. Побудуйте stackplot для 3 серій як “накопичення” в часі.

```python
x = np.arange(1, 11)
s1 = np.array([1,2,2,3,3,4,4,4,5,5])
s2 = np.array([1,1,2,2,3,3,2,2,2,1])
s3 = np.array([0,1,1,1,2,2,3,3,2,2])

fig, ax = plt.subplots(figsize=(6, 3))
ax.stackplot(x, s1, s2, s3, labels=["S1","S2","S3"], alpha=0.7)
ax.set_title("stackplot")
ax.set_xlabel("t")
ax.set_ylabel("Сума")
ax.legend(loc="upper left")
plt.show()
```

9.Побудуйте точки з вертикальними похибками.

```python
x = np.arange(1, 9)
y = np.array([2.1, 2.4, 2.9, 3.0, 3.2, 3.6, 3.8, 4.0])
err = np.array([0.2, 0.15, 0.25, 0.2, 0.18, 0.22, 0.2, 0.25])

fig, ax = plt.subplots(figsize=(6, 3))
ax.errorbar(x, y, yerr=err, fmt='o', capsize=4, linewidth=2)

ax.set_title("Errorbar")
ax.set_xlabel("Номер вимірювання")
ax.set_ylabel("Значення")
ax.grid(True, alpha=0.3)
plt.show()
```

10. Побудуйте stem plot для дискретного сигналу.

```python
x = np.arange(0, 20)
y = np.sin(0.4 * x) + 0.2 * np.cos(1.2 * x)

fig, ax = plt.subplots(figsize=(6, 3))
ax.stem(x, y)

ax.set_title("Stem plot")
ax.set_xlabel("x")
ax.set_ylabel("y[x]")
ax.grid(True, alpha=0.3)
plt.show()
```

11. Побудуйте кусково-сталий сигнал як step plot.

```python
x = np.arange(0, 11)
y = np.array([0,0,1,1,2,2,2,1,1,0,0])

fig, ax = plt.subplots(figsize=(6, 3))
ax.step(x, y, where="mid", linewidth=2)

ax.set_title("Step plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True, alpha=0.3)
plt.show()
```

12. Створіть 2×2 сітку підграфіків: line, scatter, bar, histogram.

```python
np.random.seed(5)
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)
a = np.random.randn(300)

fig, axs = plt.subplots(2, 2, figsize=(8, 5))

axs[0, 0].plot(x, y)
axs[0, 0].set_title("Line")

axs[0, 1].scatter(np.random.randn(100), np.random.randn(100), s=40, edgecolor="k")
axs[0, 1].set_title("Scatter")

axs[1, 0].bar(["A","B","C"], [3,7,5])
axs[1, 0].set_title("Bar")

axs[1, 1].hist(a, bins=20)
axs[1, 1].set_title("Histogram")

for ax in axs.ravel():
    ax.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()
```

13. Побудуйте cos(x) і підпишіть локальний максимум стрілкою.

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(t, s, linewidth=2)

ax.annotate("local max",
            xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor="black", shrink=0.05))

ax.set_title("Анотація точки")
ax.set_ylim(-2, 2)
ax.grid(True, alpha=0.3)
plt.show()
```

14. Побудуйте 3 серії даних різними маркерами та додайте legend.
